```python
from diffusers import StableDiffusionPipeline
import torch
```

```python
model_id1 = "dreamlike-art/dreamlike-diffusion-1.0"

pipe = StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float16,use_safetensors=True)
if torch.cuda.is_available():
    pipe = pipe.to("cuda")
```

```python
prompt = """A misty autumn forest with golden leaves falling, soft rain hitting the ground, and birds quietly perched on branches.
"""
```

```python
image = pipe(prompt).images[0]
```

```python
import matplotlib.pyplot as plt # Import the matplotlib library and alias it as plt

print("[PROMPT]: ",prompt)
plt.imshow(image);
plt.axis('off');
```
